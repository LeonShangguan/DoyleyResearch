```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import glob
OriginImagePath = glob.glob('../test/*.png')
OldSplitPath = glob.glob('../test/labels/*epoches.jpg')
NewSplitPath = glob.glob('../test/labels/*newsplit.jpg')
```

```python
assert len(OriginImagePath) == len(OldSplitPath) == len(NewSplitPath)
```

```python
for i in range(10):
    origin = cv2.imread(OriginImagePath[i], 0)
    old = cv2.imread(OldSplitPath[i], 0)
    new = cv2.imread(NewSplitPath[i], 0)
    plt.subplots(figsize=(16, 12))
    plt.subplot(1,3,1);plt.imshow(origin, 'gray');plt.title('origin');
    plt.subplot(1,3,2);plt.imshow(old);plt.title('old');
    plt.subplot(1,3,3);plt.imshow(new);plt.title('new');
    plt.show()
```

```python
train_loss = [0.7231, 0.3275, 0.2655, 0.404, 0.2200, 0.2066]
train_iou = [0.3463, 0.6999, 0.7564, 0.7795, 0.7982, 0.8105]
val_loss = [0.7296, 0.5775, 0.6151, 0.6798, 0.6419, 0.6270]
val_iou = [0.2688, 0.3062, 0.3287, 0.3091, 0.3019, 0.3486]
```

```python
plt.subplots(figsize=(16,5))
plt.subplot(1,2,1);plt.title('Loss');
plt.plot(train_loss, color='green', label='train_loss')
plt.plot(val_loss,  color='skyblue', label='val_loss')
plt.legend()
plt.subplot(1,2,2);plt.title('Metric');
plt.plot(train_iou, color='red', label='train_iou')
plt.plot(val_iou, color='blue', label='val_iou')
plt.legend()

plt.xlabel('Epoches');plt.show()
```
